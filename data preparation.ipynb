```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import skimage.io
import skimage.feature
import skimage.transform

import os
from glob import glob
import re
%matplotlib inline
```

```python
#read all the folders in the data

folder = os.listdir('data')
print(folder)
```

```python
all_path=[]
for f in folder:
    all_path += glob('./data/{}/*.jpg'.format(f))
    

```

```python
all_path
```

```python
#read all images and labelling

img_all = skimage.io.ImageCollection(all_path)
```

```python
len(img_all)
```

```python
img_all.files
```

```python
#labelling
#extract text folder 
def extract_label(text):
    try:
        label =re.search(r'./data/(.*?)Head\\',text).group(1)
    except:
        label =re.search(r'./data/(.*?)\\',text).group(1)
    
    return label.lower()    
```

```python
labels = list(map(extract_label,img_all.files))
```

```python
len(labels)
```

```python
def buffer(io):
    return io

img_all_arrs =np.array(list(map(buffer,img_all)))

```

```python
img_all_arrs.shape
```

```python
data = dict()

data['description'] = 'there are 20 classes and 2057 images are there. all the images are 80 X 80(rgb)'
data['data'] = img_all_arrs
data['target'] = labels
data['labels'] = set(labels)
```

```python
data
```

```python
#save the file in pickle
import pickle
import collections
```

```python
pickle.dump(data,open('data_animals_head_20.pickle','wb'))
```

```python
#visualize image and labels
```

```python
count_values = collections.Counter(data['target'])
count_values
```

```python
y_axis = list(count_values.keys())
values = count_values.values()
plt.figure(figsize=(10,6))
plt.barh(y_axis, values)
plt.xlabel('count')
plt.title('label count')
plt.show()
```

```python
plt.figure(figsize= (12,6))
for i, c in enumerate(data['labels']):
    index = data['target'].index(c)
    img = data['data'][index]
    
    plt.subplot(3,7, i+1)
    plt.imshow(img)
    plt.xticks([]), plt.yticks([])
    plt.title(c)
    
plt.show()

```
